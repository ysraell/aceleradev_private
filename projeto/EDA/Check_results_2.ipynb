```python
results_csv_name = 'Results_A2.csv'
df_laoded = pd.read_csv(results_csv_name)
df_laoded = df_laoded.loc[:, ~df_laoded.columns.str.contains('^Unnamed')]
df_laoded['P_True'] = df_laoded[['P1_True','P2_True']].sum(1)
df_laoded['P12pp'] = df_laoded[['P1pp','P2pp']].mean(1)
```
